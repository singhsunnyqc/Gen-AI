# **K-Means Clustering using ScikitLearn**

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import pandas as pd
```

```python
df.head()
```

```python
df_clustering=df.loc[:,['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
# Scale the data
scaler = StandardScaler()
df_clustering_scaled = scaler.fit_transform(df_clustering)
```

```python
# Performing K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=0)
y_kmeans = kmeans.fit_predict(df_clustering_scaled)
```

```python
# Visualizing the clusters - choose the first two features for simplicity
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_clustering_scaled[y_kmeans == 0, 0], y=df_clustering_scaled[y_kmeans == 0, 1], color='yellow', label='Cluster 1', s=50)
sns.scatterplot(x=df_clustering_scaled[y_kmeans == 1, 0], y=df_clustering_scaled[y_kmeans == 1, 1], color='blue', label='Cluster 2', s=50)
sns.scatterplot(x=df_clustering_scaled[y_kmeans == 2, 0], y=df_clustering_scaled[y_kmeans == 2, 1], color='green', label='Cluster 3', s=50)

# Plotting the centroids of the clusters
centroids = kmeans.cluster_centers_  # Inverse transform to plot centroids in original scale
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, color='red', label='Centroids', marker='X')
plt.title('K-Means Clustering on Iris Dataset (Scaled Data)')
plt.xlabel('Sepal Length (scaled)')
plt.ylabel('Sepal Width (scaled)')
plt.legend()
plt.show()
```
